# *`IMPORT LIBRARIES`*

```python

import pandas as  pd
import numpy as np

from scipy import stats
import statsmodels.api as sm

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OrdinalEncoder
from sklearn.cluster import KMeans
```

```python
#reading the dataset
df = sns.load_dataset('penguins')
df.head()
```

```python
df.info()
```

# *`DATA OBSERVATION`*

```python
df.columns
```

```python
df.isna().sum()
```

```python
columns=['species', 'island', 'sex']
for col in columns:
   print(col)
   print(df[col].unique())
   print("------------")
```

```python
'''
unsupervised learning

k-means clustering
'''
```

# *`DATA HANDLING`*

```python
df.isnull().sum()
```

```python
#filling continuous columns null values with their respective mean
df.bill_length_mm.fillna(df.bill_length_mm.mean(),inplace=True)
df.bill_depth_mm.fillna(df.bill_depth_mm.mean(),inplace=True)
df.flipper_length_mm.fillna(df.flipper_length_mm.mean(),inplace=True)
df.body_mass_g.fillna(df.body_mass_g.mean(),inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

# *`DATA PREPROCESSING`*

```python
category_columns=['species', 'island', 'sex']

encoder= OrdinalEncoder()
df["species"]=encoder.fit_transform(df[["species"]])

df["island"]=df.island.map({"Torgersen":11,"Biscoe":12,"Dream":13})

df["sex"]=df.sex.replace({"Male":1,"Female":2})
```

```python
df.head()
```

# *`DATA VISUALIZATION`*

```python

correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,linewidths=0.2,cmap="YlGnBu")
plt.title("Correlation Matrix for palmer penguin")
plt.show()
```

# *`MACHINE LEARNING`*

```python
'''
Target variable not found
all columns are independent
hence unsupervised learning using k-means clustering
'''
```

```python
wcss = []
for i in range(1, 150):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
print(wcss)

plt.plot(range(1, 10), wcss[0:9])
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
result= KMeans(n_clusters=3).fit(df)
identify_cluster=result.fit_predict(df)
df['cluster']=identify_cluster
df.cluster.value_counts()
```

```python
df2=df.apply(pd.to_numeric, errors='coerce')
sns.clustermap(df2,cmap="RdBu", robust=True)
```
